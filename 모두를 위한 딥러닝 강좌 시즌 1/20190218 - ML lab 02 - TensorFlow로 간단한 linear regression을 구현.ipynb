Hypothesis
    H(x) = Wx + b
    
Cost Function
    cost(W, b) = 1/m sigma (i=1 to m (H(x_i) - y_i)) ** 2
    
1. 그래프를 빌드한다.
2. 세션 런을 통해서 그래프를 실행시킨다.
3. 그래프의 변수를 업데이트한다.

```python
import tensorflow as tf
```

```python
# H(x) = Wx + b

X = tf.placeholder(tf.float32, shape=[None])  # 몇 개의 데이터가 나올지 모름.
Y = tf.placeholder(tf.float32, shape=[None])

# Tensorflow; Trainable Variable
# 값을 모르므로 random_normal 으로 둠 (rank가 1)
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = X * W + b

# cost(W, b) = 1/m sigma (i=1 to m (H(x_i) - y_i)) ** 2

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
# reduce_mean : avg

# Minimize (Gradient Descent Optimizer, 지금은 마법이라고 생각하자)

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
# train을 시켜야 cost minimize가 된다.

sess = tf.Session()
# Variable를 사용하기 전, 반드시 초기화 해야 함
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X: [5], Y: [2.5]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```
